```python
% 读取 Excel 文件
filename = 'E:\Matlab\EX2025\AuxiliaryData\LFMC-gridMean-ML.xlsx';
data = readtable(filename);

% 提取需要的列
lfmcValue = data.LFMCValue;
colVOD = data.ColVOD;
rowVOD = data.RowVOD;
samplingDate = data.SamplingDate;

% 将日期转换为 datetime 格式
dates = datetime(samplingDate, 'InputFormat', 'yyyy-MM-dd'); % 根据实际格式调整

% 统计每天的数据点数量
[uniqueDates, ~, idx] = unique(dates);
counts = accumarray(idx, 1); % 统计每个日期的数据点数量

% 创建包含日期和计数的表格
dateCountTable = table(uniqueDates, counts, 'VariableNames', {'Date', 'Count'});

% 按计数降序排序
sortedTable = sortrows(dateCountTable, 'Count', 'descend');

% 提取前20天
top20Dates = sortedTable(1:min(20, height(sortedTable)), :);

% 显示结果
disp('前20天数据点最多的日期:');
disp(top20Dates);

% 导出结果到Excel
outputFilename = 'Top20_Days_With_Most_Points.xlsx';
writetable(top20Dates, outputFilename);

% 可选：导出完整数据中这些日期的所有记录
% 获取前20天的日期
selectedDates = top20Dates.Date;

% 找出这些日期对应的所有行
isSelected = ismember(dates, selectedDates);
selectedData = data(isSelected, :);

% 导出完整数据
outputDataFilename = 'Complete_Data_For_Top20_Days.xlsx';
writetable(selectedData, outputDataFilename);

disp('处理完成！');
disp(['前20天统计结果已导出到: ' outputFilename]);
disp(['完整数据已导出到: ' outputDataFilename]);
```
